```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np


im = np.array(Image.open('big_stop2.JPG'), dtype=np.uint8)

# Create figure and axes
fig,ax = plt.subplots(1)

# Display the image
ax.imshow(im)

# Create a Rectangle patch
rect1 = patches.Rectangle((155,185),50,30,linewidth=2,edgecolor='r',facecolor='none')
rect2 = patches.Rectangle((65,220),70,30,linewidth=2,edgecolor='r',facecolor='none')
# Add the patch to the Axes
ax.add_patch(rect1)
ax.add_patch(rect2)

plt.show()
```
